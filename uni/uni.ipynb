## Initialisation

```python
import pandas as pd

cwurData = pd.read_csv("cwurData.csv",sep=",")
shanghaiData = pd.read_csv("shanghaiData.csv", sep=",")
timesData = pd.read_csv("timesData.csv", sep=",")
eesData = pd.read_csv("education_expenditure_supplementary_data.csv", engine='python', sep=",")
easData = pd.read_csv("educational_attainment_supplementary_data.csv", sep=",")
sacData = pd.read_csv("school_and_country_table.csv", sep=",")

cwurData_simplified = cwurData[['world_rank','institution','country','year']]
cwurData_simplified = cwurData_simplified.rename(columns={'institution':'university_name'})
shanghaiData_simplified = shanghaiData[['world_rank','university_name','year']]
timesData_simplified = timesData[['world_rank','university_name','country','year']]
```

### 2. Quelles années sont prises en charge par chaque classement ?

```python
print("CWUR :",cwurData['year'].unique())
print("Shanghai :",shanghaiData['year'].unique())
print("Times :",timesData['year'].unique())
```

### 3. Combien d'universités sont prises en compte annuellement par chaque classement en moyenne ?

```python
print("CWUR :",cwurData['year'].value_counts().mean())
print("Shanghai :",shanghaiData['year'].value_counts().mean())
print("Times :",timesData['year'].value_counts().mean())
```

### 4. Quels sont les rankings obtenus par l'Université de Strasbourg dans chacun des 3 classements disponibles ?

```python
print("Classement CWUR :")
c_strasbourg = cwurData_simplified[cwurData_simplified['university_name'].str.contains("Strasbourg")]
print(c_strasbourg[['year','world_rank']])
print("----------")
print("Classement Shanghai :")
s_strasbourg = shanghaiData_simplified[shanghaiData_simplified['university_name'].str.contains("Strasbourg", na=False)]
print(s_strasbourg[['year','world_rank']])
print("----------")
print("Classement Times :")
t_strasbourg = timesData_simplified[timesData_simplified['university_name'].str.contains("Strasbourg")]
print(t_strasbourg[['year','world_rank']])
```
